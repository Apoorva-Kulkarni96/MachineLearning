### Gradient Descent for Logistic Regression

```python
import numpy as np

def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

def hypothesis(X, theta):
    return sigmoid(np.dot(X, theta))

def error(X, Y, theta):
    hypo = hypothesis(X, theta)
    hypo = np.clip(hypo, 1e-10, 1 - 1e-10)

    return -np.mean(Y * np.log(hypo) + (1 - Y) * np.log(1 - hypo))

def gradient(X, Y, theta):
    m = X.shape[0]
    hypo = hypothesis(X, theta)
    return np.dot(X.T, (hypo - Y)) / m

def gradient_descent(X, Y, max_iter=50, lr=0.01):
    theta = np.zeros((X.shape[1],))
    error_list = []
    theta_list = []

    for i in range(max_iter):
        err = error(X, Y, theta)
        grad = gradient(X, Y, theta)

        theta = theta - lr * grad

        error_list.append(err)
        theta_list.append(theta.copy())

    return theta_list, error_list
```

### Gradient Descent for Linear Regression 

```python
import numpy as np

def hypothesis(X, theta):
    z = np.dot(X, theta)
    return z

def gradient(X,Y, theta):
    m = X.shape[0]
    hypo = hypothesis(X, theta)
    grad = np.dot(X.T, (hypo-Y))
    return grad/m
    
def error(X,Y, theta):
    m = X.shape[0]
    hypo = hypothesis(X, theta)
    err = np.mean((hypo - Y)**2)
    return err
    
def gradient_descent(X,Y,max_iter=50, lr=0.001):
    m = X.shape[0]
    theta = np.zeros((X.shape[1],))
    error_list=[]
    theta_list=[]

    for i in range(max_iter):
        err = error(X,Y, theta)
        grad = gradient(X,Y,theta)
        theta = theta - lr * grad
        error_list.append(err)
        theta_list.append(theta.copy())
    return theta_list, error_list
        
```
